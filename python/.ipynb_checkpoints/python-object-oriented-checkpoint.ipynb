```python
#Into to object oriented programming
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return (f"{self.name} is {self.age} years old")

    def greet(self):
        print(f"Hello my name is, {self.name}!")

p1=Person("John", 25)
p1.greet()
```

```python
#Inheritance
class Animal:
    def speak(self):
        print("Animal Speaks")

class Dog(Animal):
    def speak(self):
        print("Dog Speaks")

dog = Dog()
dog.speak()
```

```python
#Abstraction
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def Shape(self):
        pass

class Circle(Shape):
    def Shape(self):
        return "rectangle"

circle = Circle()
print(circle.Shape())

#this should throw error, Abstract classes cannot be instantiated!
shape = Shape()
print(shape.Shape())
```

```python
## Assignment: Create a class shape with a method that returns the area, Inherit the shape class to Circle, Rectangle, Triangle
import math

class MyShape(ABC):
    @abstractmethod
    def area(self):
        pass

class MyTriangle(MyShape):
    def __init__(self, length, height):
        self.length = length
        self.height = height
        
    def area(self):
        return 1/2*self.length*self.height

class MyRectangle(MyShape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
        
    def area(self):
        return 1/2*self.length*self.width

class MyCircle(MyShape):
    def __init__(self, radius):
        self.radius = radius
        
    def area(self):
        return (self.radius**2)*math.pi

triangle=MyTriangle(2,2)
print(f"MyTriangle Area: {triangle.area()}")

rectangle=MyRectangle(2,2)
print(f"MyRectangle Area: {rectangle.area()}")

circle=MyCircle(2)
print(f"MyCircle Area: {circle.area()}")
```

```python
#Super call
class MyAnimal: 
    def __init__(self, name):
        self.name = name
    def speak(self):
        print(f"{self.name} makes a sound")

class Cat(MyAnimal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color

    def describe(self):
        print(f"{self.name} is a {self.color} cat")

cat = Cat("cat", "blue")
cat.describe()
```
