```python
# 2. Given a string s, you need to check if it is palindrome or not. A palidrome is a string that reads the same from front and back.
def isPalidrome(word):
    l = 0
    r = len(word)-1
    while(l<r):
        if(word[l]!=word[r]):
            return False
        l+=1
        r-=1
    return True

print(isPalidrome("racecar"))
```

```python
# 3. Given a string s, and a pattern p. You need to find if p exists in s or not and return the starting index of p in s. If p does not exist in s then -1 will be returned.
# Here p and s both are case-sensitive.
 
# Examples:
# Input: s = "Hello", p = "llo"
# Output: 2
# Explanation: llo starts from the second index in Hello.

def strContains(string, pattern):
    index = 0
    start = -1
    patLen = len(pattern)
    for i in range(len(string)):
        if index == patLen:
            break
            
        if string[i] == pattern[index]:
            if start == -1:
                start = i 
            index+=1;
        else:
            start = -1
            index = 0
    return start
    
print(strContains("Helloaooollo", "llo"))
```

```python
# Dictionary
# 1.Given an array arr[], find the first repeating element. The element should occur more than once and the index of its first occurrence should be the smallest.
# Examples:
# Input: arr[] = [1, 5, 3, 4, 3, 5, 6]
# Output: 2
# Explanation: 5 appears twice and its first appearance is at index 2 which is less than 3 whose first the occurring index is 3.

def firstRepeatingValue(arr):
    dic = {}
    for i in range(len(arr)):
        if arr[i] in dic:
            return dic[arr[i]]
        else:
            dic[arr[i]] = i
    return -1

print(firstRepeatingValue([1, 5, 3, 4, 3, 5, 6]))
```

```python
# Set
# 1. You are given an array arr[] of size n. You have to insert all elements of arr[] into a set and return that set.
#You are also given a interger x. If x is found in set then erase it from set and print "erased x", otherwise, print "not found".

def setFunct(arr, x):
    mySet = set()
    for item in arr:
        mySet.add(item)
    if x in mySet:
        mySet.remove(x)
        print(f"erased {x}")
    else:
        print("not found")
    return mySet

print(setFunct([1,1,2,3,3,4,4,5,5], 5))    
```
