```python
"""
File handling:
1) Write script to copy 1 file contents to another file
"""
with open("file1.txt", "w") as file:
    file.write("Copy this text!")

with open("file2.txt", "w") as file2:
    with open("file1.txt", "r") as file1:
        content = file1.read()
        file2.write(content)

with open("file2.txt", "r") as file:
    print(file.read())
    
```

```python
""" 
2) Write script to copy an image file's contents to another file
"""
with open("horse_copy.jpeg", "wb") as paste:
    with open("horse.jpeg", "rb") as copy:
        image = copy.read()
        paste.write(image)
```

```python
"""
3) Write script(s) to receive information from console input to create a dictionary, then append to a file. Then, be able to read from the
file and search for relevant data to return. Start with name search then can go to any data search.
"""
import json

def append_dictionary(file_name):
    dictionary = {}
    while(True):
        pair = input("Enter dictionary Item in fromat(key:value) or Enter STOP to exit: ")
        if(pair == "STOP"): break
        key, value = pair.split(":")
        dictionary[key] = value  
    with open(file_name, "a") as file:
        for key,value in dictionary.items():
            file.write(key+":"+value+"\n")

def read_file(search_key, file_name):
    temp = {}
    with open(file_name, "r") as file:
        with open("dump.txt", "a") as dump:             
            for line in file:
                if search_key in line:
                    key, value = line.split(":")
                    temp[key] = value
                    json.dump(temp, dump)             
file = "dict.txt"
append_dictionary(file)
print("Append Successful!")
read_file("frog", file)
print("Read Successful!")
```
