```python
"""
To-Do List Manager: Create a program to add, view, mark as complete, and delete tasks from a list/dictionary, 
potentially saving the list/dictionary to a file. Please use the following as well while writing code
1. Functions
2. Modules (Optional)
3. Imports (Optional)
"""
import json

def add_tasks(file_name):
    dictionary = {}
    while(True):
        task = input("Enter Task_Name or Enter STOP to exit: ")
        if(task == "STOP"): break
        dictionary[task] = "Incomplete"  
    with open(file_name, "a") as file:
        for task, status in dictionary.items():
            file.write(task+":"+status+"\n")

def view_tasks(file_name):
    print("\nViewing Tasks:")
    with open(file_name, "r") as file:
        for line in file:
            if not line.strip():
                continue
            print(line,end="")   

def mark_task(mark, file_name):
    temp = {}
    with open(file_name, "r") as file:
        for line in file:
            if not line.strip():
                continue
            task, status = line.split(":")
            if task == mark:
                status = "Complete"
            temp[task] = status
        
    with open(file_name, "w") as file:                 
        for task, status in temp.items():
            file.write(task+":"+status+"\n")


def delete_task(delete, file_name):
    temp = {}
    with open(file_name, "r") as file:
        for line in file:
            if not line.strip():
                continue
            task, status = line.split(":")
            if task == delete:
                continue
            temp[task] = status
        
    with open(file_name, "w") as file:                 
        for task, status in temp.items():
            file.write(task+":"+status+"\n")          
```

```python
def controller():
    file = "tasks.txt"
    while(True):
        view_tasks(file)
        mode = input("Enter Mode: 1=add_tasks, 2=mark_task, 3=delete_task, 4=exit")
        match mode:
            case "1":
               add_tasks(file) 
            case "2":
                mark = input("Enter task to mark complete:")
                mark_task(mark, file)
            case "3":
                delete = input("Enter task to delete complete:")
                delete_task(delete,file)
            case "4":
                break
            case _:
                print("Enter valid mode!")

controller()
```
