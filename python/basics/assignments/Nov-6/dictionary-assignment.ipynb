```python
"""
1. Create a dictionary of five countries and their capitals. Write a function that takes a country name as input and returns its capital.
"""
capitals = {"Afghanistan": "kabul", "Albania": "Tirana", "Austria":"Canberra", "Brazil": "Brasilia", "Belarus":"Minsk"}
def find_capital(country):
    return capitals[country]

print(find_capital("Brazil"))
```

```python
"""
2. Make a dictionary of student names and their scores. Write a function to find the student with the highest score.
"""
students = {"Alice": 85,"Bob": 92,"Charlie": 78,"David": 95,"Emma": 88}
def find_top_student(student_dict):    
    student = max(student_dict, key=student_dict.get)
    score = student_dict[student]
    return student, score

print(find_top_student(students))
```

```python
"""
3. Create a nested dictionary of three employees, each with keys for name, age, and salary. Write a function to give each employee a 10% raise
and print the updated dictionary.
"""
employees = {
    1: {"name":"David","age":60, "salary":10},
    2: {"name":"Alice","age":45, "salary":10},
    3: {"name":"Bob","age":30, "salary":10}
}

def give_raise(list_employees):
    for key, value in list_employees.items():
        value["salary"] = value["salary"] * 1.1
    print(list_employees)    

give_raise(employees)
```

```python
"""
4. Write a Python program to add a key to a dictionary
Sample Output
dictionary = {"Name" : "Ram" , "Age" : 23}
add_key = {"City" : "Salem"}
dictionary = {'Name' : 'Ram', 'Age' : 23, 'City' : 'Salem'}
"""
def add_key_value(dic, key_value):
    for key, value in key_value.items():
        dic[key] = value
    print(dic)
dictionary = {"Name" : "Ram" , "Age" : 23}
add_key = {"City" : "Salem"}
add_key_value(dictionary, add_key)
```

```python
"""
5. Write a Python program to concatenate following dictionaries to create a new one.
Sample Output
Dictionary 1 = {"Name" : "Ram" , "Age" : 23}
Dictionary 2 = {"City" : "Salem", "Gender" : "Male"}
Concatenate Dictionaries = {'Name' : 'Ram', 'Age' : 23, 'City' : 'Salem', 'Gender': 'Male'}
"""
dict1= {"Name" : "Ram" , "Age" : 23}
dict2 = {"City" : "Salem", "Gender" : "Male"}
def combine_dict(dict1, dict2):
    return {**dict1, **dict2}
print(combine_dict(dict1,dict2))
```

```python
"""
6. Write a Python program to check whether a given key already exists in a dictionary.
Sample Output
{'Name' : 'Ram', 'Age' : 23,}
Key = Name
Key is Available in the Dictionary
"""
def exists(dic, key):
    if (dic.get(key) != None):
        return True
    return False

dic = {'Name' : 'Ram', 'Age' : 23,}
key = "Name"

print(exists(dic,key))
```

```python
"""
7. Write a Python program to iterate over dictionaries using for loops.
Sample Output
{"Name" : "Ram" , "Age" : 23 , "City" : "Salem", "Gender" : "Male"}
Name : Ram
Age : 23
City : Salem
Gender : Male
"""
def iterate(dic):
    for key, value in dic.items():
        print(str(key)+": "+str(value))
iterate({"Name" : "Ram" , "Age" : 23 , "City" : "Salem", "Gender" : "Male"})
```
