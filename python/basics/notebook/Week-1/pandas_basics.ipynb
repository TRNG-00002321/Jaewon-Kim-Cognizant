```python
"""
What is Pandas?
Pandas is a Python library used for working with data sets. It has functions for analyzing, cleaning, exploring, and manipulating data. The name
"Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

Why Use Pandas?
Pandas allows us to analyze big data and make conclusions based on statistical theories. Pandas can clean messy data sets, and make them readable
and relevant. Relevant data is very important in data science.
"""
import pandas as pd
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}
myvar = pd.DataFrame(mydataset)
print(myvar)
```

```python
"""
Pandas Series:
A Pandas Series is like a column in a table. It is a one-dimensional array holding data of any type.
"""
#Creating a simple Pandas Series from a list:
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
print("\n")

""" 
Labels: If notthing is specified, the vlaues are labeled with their index number like seen above. You can also name your own labels
"""
a = [1,7,2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
print("\n")

"""Key/value object can also be used when creating a series"""
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

```python
"""
Dataframe: Pandas DataFrame is a 2 dimensional datastructure, like a 2-D array or a table with row and columns
"""
#Creating a simple dataframe
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
print(df)
print("\n")
#You can locate one or more specified rows(s) row with .loc[] on the dataframe
print(df.loc[0])
print("\n")

"""
Like series you can lable the indexes with your own values
"""
df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print(df)
print("\n")

#you can use .loc[] with the custom named indexes
print(df.loc["day2"])
```

```python
"""
Pandas Read CSV: A simple way to store big data sets is to use CSV files (comma separated files). CSV files contains plain text and is a well
know format that can be read by everyone including Pandas.
"""
#Read CSV, NOTE: to_string() will only print the first and last 5 row for large data sets
df = pd.read_csv('data.csv')
print(df.to_string())
print("\n")
#the number of rows is defined in pandas settings and can be checked and changed with the following:
pd.options.display.max_rows = 60
print("Max Rows:",pd.options.display.max_rows)
df = pd.read_csv('data.csv')
print(df) 
```

```python
""" 
Pandas can also be used to read json (JSON is plain text, but has the format of an object)
"""
df = pd.read_json('data.json')
#print(df.to_string()) to print full file!
print(df) 

"""
JSON objects have the same format as Python dictionaries.If your JSON code is not in a file, but in a Python Dictionary, you can load 
it into a DataFrame directly:
"""
data = {
  "Duration":{
    "0":60,
    "1":60,
    "2":60,
    "3":45,
    "4":45,
    "5":60
  },
  "Pulse":{
    "0":110,
    "1":117,
    "2":103,
    "3":109,
    "4":117,
    "5":102
  },
  "Maxpulse":{
    "0":130,
    "1":145,
    "2":135,
    "3":175,
    "4":148,
    "5":127
  },
  "Calories":{
    "0":409,
    "1":479,
    "2":340,
    "3":282,
    "4":406,
    "5":300
  }
}

df = pd.DataFrame(data)

print(df) 
```

```python
""" 
Pandas - Analyzing DataFrames:
"""
#Use the .head(n) method to get the first n elements of a dataframe
df = pd.read_csv('data.csv')
print(df.head(10))
print("\n")

#Use the .tail(n) to get the last n elements of a dataframe
df = pd.read_csv('data.csv')
print(df.tail(10))
```

```python
"""
Pandas - Cleaning Data:
Data cleaning means fixing bad data in your data set.
Bad data could be:
-Empty cells
-Data in wrong format
-Wrong data
-Duplicates
"""
#Use .dropna() to drop all rows with empty/null cells for a data set NOTE:dropna() returns a new dataframe and doesnt affect the original!
df = pd.read_csv('data.csv')
new_df = df.dropna()
print(new_df.to_string())
print("\n")

#.duplicated() method returns true for rows that are duplicates and false for roles that are not
print(df.duplicated())
print("\n")

#To remove duplicates use .drop_duplicates(inplace=true)
df.drop_duplicates(inplace = True)
print(df)
```
