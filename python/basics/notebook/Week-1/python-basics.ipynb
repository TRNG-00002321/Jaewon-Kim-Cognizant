```python
#Python string interning: Strings are resused when possible. EX: s1 = "bob" s2 = "bob", s2 and s1 point to the same mem address
import sys

# Automatic interning for simple literals
str1 = "hello"
str2 = "hello"
print(str1 is str2)  # Output: True

# Strings with spaces are not automatically interned
str3 = "hello world"
str4 = "hello world"
print(str3 is str4)  # Output: False

# Explicit interning using sys.intern()
interned_str3 = sys.intern(str3)
interned_str4 = sys.intern(str4)
print(interned_str3 is interned_str4)  # Output: True
```

```python
#List comprehension
#Activity: Create a list containing 5 elements (anything). 
#Based on this list, create another list having elements from the previous list containing the character 'a'
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
char = 'a'

def filterList(lis, char):
    newList = []
    for item in lis:
        if char in item:
            newList.append(item)
    return newList
print(filterList(fruits, char))

#Using list comperhension 
newList = [x for x in fruits if char in x]
print(newList)

#another example
numbers = [1, 2, 3, 4, 5, 6]
# Structure:   [ Output: num | Collection: for num in numbers | condition: if num % 2 == 0 ]
even_numbers = [num for num in numbers if num % 2 == 0]
print(even_numbers)
```

```python
#Exceptions
#Write a python program to input two numbers. Calculate and display their division.
def div():
    try:
        a = input("enter numerator:")
        b = input("enter denominator")
        print(int(a)/int(b))
    except ZeroDivisionError:
        print("except ZeroDivisionError: Cannot divide by zero!")
    except ValueError:
        print("except ValueError: Must divide two numbers!")
    except TypeError:
        print("except TypeError: Must divide two numbers!")
    else:
        print("else: is ran if no errors/exceptions here")
    finally:
        print("finally: always run; error or no error!")
div()
```

```python
#Garbage collection: 

#Python uses reference counting to manage memory. Each object keeps track of how many references point to it. 
#When the reference count drops to zero i.e., no references remain, Python automatically deallocates the object
import sys

x = [1, 2, 3]
print(sys.getrefcount(x)) 
y = x
print(sys.getrefcount(x)) 
y = None
print(sys.getrefcount(x))
```

```python
#Problem with Reference Counting: Reference counting fails in the presence of cyclic references i.e., objects that reference each other in a cycle.
#Even if nothing else points to them, their reference count never reaches zero. Example:
x = [1, 2, 3]
y = [4, 5, 6]

x.append(y)
y.append(x)
print(sys.getrefcount(x))
print(sys.getrefcount(y))

#Python’s Generational Garbage Collector is designed to deal with cyclic references. 
#It organizes objects into three generations based on their lifespan:
    #-Generation 0: Newly created objects.
    #-Generation 1: Objects that survived one collection cycle.
    #-Generation 2: Long-lived objects.
```

```python
#When reference cycles occur, the garbage collector automatically detects and cleans them up, freeing the memory.
#Garbage collection runs automatically when the number of allocations exceeds the number of deallocations by a certain threshold. 
#This threshold can be inspected using the gc module.

import gc
print(gc.get_threshold())
#It returns the threshold tuple for generations 0, 1 and 2. When allocations exceed the threshold, collection is triggered.
```

```python
#Sometimes it’s beneficial to manually invoke the garbage collector, especially in the case of reference cycles. Example:
# Create a cycle
def fun(i):
    x = {}
    x[i + 1] = x
    return x

# Trigger garbage collection
c = gc.collect()
print(c)

for i in range(10):
    fun(i)

c = gc.collect()
print(c)
#def fun(i) creates a cyclic reference by making a dictionary reference itself.
#gc.collect() triggers garbage collection and stores the count of collected objects (initially 0).
#for i in range(10) calls fun(i) 10 times, creating 10 cyclic references.
#gc.collect() triggers garbage collection again and prints the count of collected cycles.
```
