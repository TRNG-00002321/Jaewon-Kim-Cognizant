```python
#Python string interning: Strings are resused when possible. EX: s1 = "bob" s2 = "bob", s2 and s1 point to the same mem address
import sys

# Automatic interning for simple literals
str1 = "hello"
str2 = "hello"
print(str1 is str2)  # Output: True

# Strings with spaces are not automatically interned
str3 = "hello world"
str4 = "hello world"
print(str3 is str4)  # Output: False

# Explicit interning using sys.intern()
interned_str3 = sys.intern(str3)
interned_str4 = sys.intern(str4)
print(interned_str3 is interned_str4)  # Output: True
```
